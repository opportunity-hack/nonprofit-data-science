```python
import glob
import re
file_location = f"{mydir}/*.txt"
read_files = glob.glob(file_location)


all_text = []
for f in read_files:
    print(f"Reading from {f}...")
    with open(f, "r") as infile:     
        text = infile.read()
        text = text.replace("\n", " ").replace("\t", " ")
        text = re.sub("\s\s+", " ", text)
        all_text.append(text)

giant_text_string = " ".join(all_text)
```

```python
import spacy
#spacy.load('en')
from spacy.lang.en import English
parser = English()
def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
    
from nltk.stem.wordnet import WordNetLemmatizer
def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

```python
nltk.download('stopwords')
en_stop = set(nltk.corpus.stopwords.words('english'))
```

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in en_stop]
    tokens = [get_lemma(token) for token in tokens]
    return tokens
```

```python
import random
text_data = []
for line in all_text:
    tokens = prepare_text_for_lda(line)
    text_data.append(tokens)
    # If you need to sample your data because there is too much, you can use this:
    #if random.random() > .99:
    #    print(tokens)
    #    text_data.append(tokens)
```

```python
word_flat_list = [item for sublist in text_data for item in sublist]
```

```python
word_flat_list
```

```python
from collections import Counter
c = Counter(word_flat_list)
most_occur = c.most_common(20)
print(most_occur)
```

```python
from nltk import ngrams
ngram_counts = Counter(ngrams(word_flat_list, 10))
ngram_counts.most_common(10)
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]
import pickle
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

```python
import gensim
NUM_TOPICS = 15
# LDA, or Latent Dirchlet Allocation is arguably the most famous topic modeling algorithm out there. 
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
ldamodel.save('model5.gensim')
#ldamodel.show_topics()
topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```

```python
# HDP, the Hierarchical Drichlet Process is an unsupervised topic model which figures out the number of topics on it's own.
hdp_model = gensim.models.hdpmodel.HdpModel(corpus=corpus, id2word=dictionary)
hdp_model.save('model.hdp.gensim')
# hdp_model.show_topics()
topics = hdp_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```
